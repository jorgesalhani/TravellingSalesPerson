```python
import matplotlib.pyplot as plt
import numpy as np

f = open("executionTime.txt", 'r')
conexoes = []
t = []
i = 0 
for l in f.readlines():
  data_i = l.strip().split(',')
  if i > 0:
    conexoes.append(int(data_i[0]))
    t.append(float(data_i[1]))
  i+=1
```

```python
def factorial(n):
  if (n == 0 or n == 1): return 1
  return n * factorial(n-1)
```

```python
x = np.linspace(20,68,1000)
y = [np.exp(i/20)-3 for i in x]
```

```python
fig, ax = plt.subplots()
ax.plot(conexoes[1:], t[1:], 'ko', label='Experimental')
# ax.plot(x,y, 'r-', label='Experimental')
ax.set_xlabel('Número de Conexões (N)')
ax.set_ylabel('Tempo (s)')
ax.legend()
plt.show()
```
